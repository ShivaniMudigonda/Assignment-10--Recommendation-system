```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('book (1).csv', encoding='unicode escape')
df
```

```python
df1 = df.drop(['Unnamed: 0'], axis = 1)
df1
```

```python
df1.columns = ['userId','BookTitle','BookRating']
df1
```

```python
#number of unique users in the dataset
len(df1.userId.unique())
```

```python
len(df1.BookTitle.unique())
```

```python
user_df1 = df1.pivot_table(index='userId',
                                 columns='BookTitle',
                                 values='BookRating').reset_index(drop=True)
user_df1
```

```python
user_df1.index = df1.userId.unique()
user_df1
```

```python
#Impute those NaNs with 0 values
user_df1.fillna(0, inplace=True)
user_df1
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_df1.values,metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = df1.userId.unique()
user_sim_df.columns = df1.userId.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
